```python
%matplotlib inline
from nuscenes.nuscenes import NuScenes

nusc = NuScenes(version='v1.0-mini', dataroot='/data/sets/nuscenes', verbose=True)
```

```python
nusc.list_scenes()
```

```python
my_scene = nusc.scene[0]
my_scene
```

```python
first_sample_token = my_scene['first_sample_token']
```

```python
my_sample = nusc.get('sample', first_sample_token)
my_sample
```

```python
my_sample['token']
```

```python
list_token = []

for i in range(10):
    my_scene = nusc.scene[i]
    first_sample_token = my_scene['first_sample_token']

    current_token = first_sample_token

    while current_token:
        my_sample = nusc.get('sample', f"{current_token}")
        list_token.append(my_sample['token'])
        current_token = my_sample['next']


list_token
```
